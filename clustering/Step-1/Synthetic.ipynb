```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
```

```python
sc = SparkContext()
spark = SparkSession(sc)
sqlContext = SQLContext(sc)
```

```python
infile = "s3://srifinal007/sample_geo.txt"
```

```python
d1 = sc.textFile(infile)
d2 = d1.filter(lambda x: (x!="") and ("L" not in x) )
d3 = d2.map(lambda x: x.split("\t"))
d4 = d3.map(lambda x: (float(x[0]),float(x[1]),int(x[2])))
```

```python
actdata = d4.collect()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(actdata, columns=["Latitude", "Longitude", "IndexID"])
df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6))
plt.scatter(df.Latitude,df.Longitude, s = 7, c = "blue", marker = "X", alpha = 0.25, linewidths = 1)
```
